#**Visualize the data using seabborn**

##**Scatter Plot**

```python
sb.pairplot(data)
```

##**Histogram Plot**

```python
plt.figure(figsize=(15,5))
plt.subplot(1,4,1)
sb.histplot(df['sepal_length'] ,bins=20)
plt.title('sepal_length')
plt.subplot(1,4,2)
sb.histplot(df['sepal_width'] ,bins=20)
plt.title('sepal_width')
plt.subplot(1,4,3)
sb.histplot(df['petal_length'] ,bins=20)
plt.title('petal_length')
plt.subplot(1,4,4)
sb.histplot(df['petal_width'] ,bins=20)
plt.title('petal_width')
plt.subplots_adjust(wspace=0.3)
plt.show()
```

##**Box Plot**

```python
plt.figure(figsize=(15,5))
plt.subplot(1,4,1)
sb.boxplot(df['sepal_length'])
plt.title('sepal_length')
plt.subplot(1,4,2)
sb.boxplot(df['sepal_width'])
plt.title('sepal_width')
plt.subplot(1,4,3)
sb.boxplot(df['petal_length'])
plt.title('petal_length')
plt.subplot(1,4,4)
sb.boxplot(df['petal_width'])
plt.title('petal_width')
plt.subplots_adjust(wspace=0.3)
```

###**Explanation:**
###Plotted the graphs using seaborn library to visualise the data in a bettery way.Three graphs were plotted:
###**Scatter plot:**
###It represents the relationship between the 2 variables present in the data. As can see above there are 16 scatter plots, each plot is shwoing the spread of the values between 2 columns/variable in the data.
###**Histogram**
###It represents the distribution of the numerical values of each column in the data.Data is distributed in multiple bins for example second plot(sepal_width) has value range from 2 to 4 half values of sepal width lie between 2 & 3 and half values lie betwenn 3 & 4. Same for others.
###**Box plot**
###Box plot is used to see the outliers in the data. Above four plots were plotted, as can see the second plot (sepal_width) has few value which is lying beyound the top wisker of box plot there, so these are outliers. Rest all the box plots has no extreme values/outliers.

#**Visualize the data using matplotlib**

##**Scatter Plot**

```python
[col for col in data.columns]
```

```python
l1=[col for col in df.columns]
l1[1:-1]
```

```python
l = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']

plt.figure(figsize=(15,10))
ax = 0

for i in range(len(l)):
    for j in range(len(l)):
        ax += 1
        plt.subplot(4, 4, ax)
        plt.scatter(df[l[i]], df[l[j]])
        plt.xlabel(l[i])
        plt.ylabel(l[j])

plt.subplots_adjust(hspace=0.5, wspace=0.5)
plt.show()
```

##**Histogram Plot**

```python
plt.figure(figsize=(15,5))
plt.subplot(1,4,1)
plt.hist(df['sepal_length'] ,bins=20)
plt.title('sepal_length')
plt.subplot(1,4,2)
plt.hist(df['sepal_width'] ,bins=20)
plt.title('sepal_width')
plt.subplot(1,4,3)
plt.hist(df['petal_length'] ,bins=20)
plt.title('petal_length')
plt.subplot(1,4,4)
plt.hist(df['petal_width'] ,bins=20)
plt.title('petal_width')
plt.show()
```

##**Box Plot**

```python
l=[col for col in df.columns]
plt.figure(figsize=(15,10))
ax=0
for col in range(len(l)-1):
    ax+=1
    plt.subplot(2,3,ax)
    plt.boxplot(df[l[col]], patch_artist=True)
    plt.title(l[col])

```

###**Explanation:**
###Plotted the graphs using matplotlib library to visualise the data in a bettery way.Three graphs were plotted:
###**Scatter plot:**
###It represents the relationship between the 2 variables present in the data. As can see above there are 16 scatter plots, each plot is shwoing the spread of the values between 2 columns/variable in the data.
###**Histogram**
###It represents the distribution of the numerical values of each column in the data.Data is distributed in multiple bins for example second plot(sepal_width) has value range from 2 to 4 half values of sepal width lie between 2 & 3 and half values lie betwenn 3 & 4. Same for others.
###**Box plot**
###Box plot is used to see the outliers in the data. Above four plots were plotted, as can see the second plot (sepal_width) has few value which is lying beyound the top wisker of box plot there, so these are outliers. Rest all the box plots has no extreme values/outliers.
